# Linear Regression

```python
%matplotlib inline
import numpy as np
from pylab import *

Grades = np.random.normal(3.0,1.0,1000)
Effort = 100 - (Grades + np.random.normal(0, 0.1, 1000)) * 3

scatter (Grades,Effort)
```

```python
from scipy import stats

slope , intercept , r_value , p_value, std_err = stats.linregress(Grades,Effort)
```

```python
r_value ** 2
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(Grades)

plt.scatter(Grades,Effort)
plt.plot(Grades , fitLine, c='r')
plt.show()
```
